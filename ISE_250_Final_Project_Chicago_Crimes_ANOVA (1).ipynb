```python
# prompt: Do a chi-square test between primary type and location description

import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
contingency_table = pd.crosstab(sample_df['Primary Type'], sample_df['Location Description'])

# Perform the chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
print(f"Degrees of freedom: {dof}")
print("Expected frequencies:\n", expected)
```

##The Impact Of Demographics On Crime Rate:

```python
import statsmodels.formula.api as smf
import statsmodels.api as sm  # This is where anova_lm lives

# Fit the model
model_for_youth = smf.ols('Q("Total Crimes Per Capita") ~ pct_youth', data=df_for_analysis).fit()

# Print model summary
print(model_for_youth.summary())

# Perform Type II ANOVA
anova_table_youth = sm.stats.anova_lm(model_for_youth, typ=2)
print(anova_table_youth)
```

```python
model_for_young_adult = smf.ols('Q("Total Crimes Per Capita") ~ pct_young_adult', data=df_for_analysis).fit()
print(model_for_young_adult.summary())
anova_table_young_adult = sm.stats.anova_lm(model_for_young_adult, typ=2)
print(anova_table_young_adult)
```

```python
model_for_middle_age = smf.ols('Q("Total Crimes Per Capita") ~ pct_middle_age', data=df_for_analysis).fit()
print(model_for_middle_age.summary())
anova_table_middle_age = sm.stats.anova_lm(model_for_middle_age, typ=2)
print(anova_table_middle_age)
```

```python
model_for_senior = smf.ols('Q("Total Crimes Per Capita") ~ pct_senior', data=df_for_analysis).fit()
print(model_for_senior.summary())
anova_table_senior = sm.stats.anova_lm(model_for_senior, typ=2)
print(anova_table_senior)
```

```python
model_for_collective_demographic = smf.ols('Q("Total Crimes Per Capita") ~ pct_youth + pct_young_adult + pct_middle_age', data=df_for_analysis).fit() #To avoid collinearity, we exclude pct_senior
print(model_for_collective_demographic.summary())
anova_table_collective_demographic = sm.stats.anova_lm(model_for_collective_demographic, typ=2)
print(anova_table_collective_demographic)
```

```python
# Fit the model
model_for_native = smf.ols('Q("Total Crimes Per Capita") ~ FOR_BORN', data=df_for_analysis).fit()

# Print model summary
print(model_for_native.summary())

# Perform Type II ANOVA
anova_table_for_native = sm.stats.anova_lm(model_for_native, typ=2)
print(anova_table_for_native)
```

```python
# Fit the model
model_for_education = smf.ols('Q("Total Crimes Per Capita") ~ assoc_plus_pct', data=df_for_analysis).fit()

# Print model summary
print(model_for_education.summary())

# Perform Type II ANOVA
anova_table_for_education = sm.stats.anova_lm(model_for_education, typ=2)
print(anova_table_for_education)
```
